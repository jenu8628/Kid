```python
import pandas as pd
csv_test = pd.read_csv('./data/foodingredient.csv')
```

```python
for i in range(0, len(csv_test)):
    id = "foodingredient-" + str(i + 1)
    csv_test.loc[i].foodingredient_id = id
```

```python
csv_test.set_index('foodingredient_id')
```

```python
csv_test.to_csv("./data/foodingredient.csv",sep=",")
```
