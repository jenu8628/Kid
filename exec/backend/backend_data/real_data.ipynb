##### random function

```python
def rand(size, random_size):
    list = []
    rand_num = random.randint(1, size)
    
    for i in range(random_size):
        while rand_num in list:
            rand_num = random.randint(1, size)
        list.append(rand_num)
    
    list.sort()
    return list
```

## Ratings Dataframe

```python
def makeRatings() :
    ratings = pd.DataFrame(columns=['recipe_id', 'recipe_name', 'user_id', 'rating'])
    
    for i in range(0, len(preferIngredients)):
        ingredient_name = preferIngredients.iloc[i]['foodingredient_name']
        prefer_sample = recipe[recipe['recipe_ingredient'].str.contains(ingredient_name, na=False)]

        for j in range(0, len(prefer_sample)):
            prefer_recipeId = prefer_sample.iloc[j]['recipe_id']
            prefer_recipeName = prefer_sample.iloc[j]['recipe_name']
            prefer_userId = preferIngredients.iloc[i]['user_id']
            prefer_rating = 4 

            check = ratings.loc[(ratings['user_id'] == prefer_userId) & (ratings['recipe_id'] == prefer_recipeId)]
            if len(check) < 1 :
                ratings = ratings.append({'recipe_id' : prefer_recipeId, 'recipe_name' : prefer_recipeName, 'user_id' : prefer_userId, 'rating' : prefer_rating}, ignore_index=True)

    return ratings
```
