```python
import sys
import numpy as np
import pandas as pd
import re
import konlpy
```

```python
def returnNewStr(inputStr):
    temp = inputStr.split(',')
    new_str = ''
    for i in range(0, len(temp)):
        str_temp = temp[i]
        sp_temp = str_temp.split(' ')
        for idx in range(0, len(sp_temp)):
            if not (any(chr.isdigit() for chr in sp_temp[idx])): # 숫자 미포함 
                if sp_temp[idx] is not '':
                    new_str += sp_temp[idx] + '/'
    new_str = new_str[0:-1]
    return new_str
```

```python
new_ingredientList = []

for ingredient_idx in range(0, len(recipe_ingredient)):
    new_str = returnNewStr(recipe_ingredient[ingredient_idx])
    new_ingredientList.append(new_str)
```

```python
ingredient_array = np.array(new_ingredientList)

# type(ingredient_array)
```

```python
#텍스트 정제 (형태소 추출)
recipeIngredient_list = []

for i, document in enumerate(ingredient_array):
    okt = konlpy.tag.Okt()
    clean_words = []
    for word in okt.pos(document, stem=True): # 어간 추출
        # 조사, 어미, 구두점, 형용사, 동사 제외
        if word[1] not in ['Josa', 'Eomi', 'Punctuation', 'Adjective', 'Verb']: 
            clean_words.append(word[0])

    recipeIngredient_list.append(clean_words)
    document = ' '.join(clean_words)

    recipe_data[i] = document

# recipeIngredient_list
```
